## View 100_random_fed results

These results were generated by the following process: 
1. Identify which cases went to trial using CaseDirectory.write_metadata.
2. Generate LLM prompts to classify the ruling of each trial using VariableExtractor.test_context
3. Give each prompt to llama3.1 70b using [Perplexity Playground](https://labs.perplexity.ai/)

- Each row in fed_results includes case metadata, the llm prompts generated from the docket_report and the documents, and the category generated by the llm. 
- fed_logs contains the log file recording how the llm prompts for each document were generated. The i'th entry of fed_logs corresponds to the i'th row of fed_results

```python
fed_results = pd.read_csv("output/100_random_fed/fed_results.csv")
```

```python
with open("output/100_random_fed/fed_logs.pkl", "rb") as f:
    fed_logs = dill.load(f)
```
